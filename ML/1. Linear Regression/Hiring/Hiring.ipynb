```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
from word2number import w2n
```

```python
df=pd.read_csv('hiring.csv')
df
```

```python
df.iloc[:,0]=df.iloc[:,0].fillna('zero')
df
```

```python
experience_num=df.iloc[:,0].apply(w2n.word_to_num)
experience_num
```

```python
df.iloc[:,0]=experience_num
df
```

```python
mean_test_score=round(df.iloc[:,1].mean(),1)
mean_test_score
```

```python
df.iloc[:,1]=df.iloc[:,1].fillna(mean_test_score)
df
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['experience','test_score(out of 10)','interview_score(out of 10)']],df['salary($)'])
```

```python
test_arr=np.array([[2,9,9],[5,6,7]])
reg.predict(test_arr)
```
