```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
df.shape
year=df['year'].values.reshape(-1,1)
income=df['per capita income (US$)']
```

```python
plt.scatter(year,income)
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(year,income)
```

```python
reg.predict([[2020]])
```

```python
test_years=np.arange(1970,2040,1).reshape(-1,1)
```

```python
test_income=reg.predict(test_years)
```

```python
plt.xlabel('Year')
plt.ylabel('Per Capita Income ($)')
plt.plot(test_years,test_income)
plt.scatter(year,income)
plt.show()
```
