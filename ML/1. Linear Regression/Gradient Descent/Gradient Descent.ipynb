```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.read_csv('test_scores.csv')
df
```

```python
def gradient_descent(x,y):
    m=0
    b=0
    iterations=1000000
    learning_rate=0.0002
    tol=1e-20
    n=len(x)
    cost=0
    for i in range(iterations):
        y_predict=m*x+b
        cost_new=1/(n)*sum((y-y_predict)**2)
        print(f"m:{m}, b:{b}, cost:{cost_new}, iteration:{i} ")
        if np.isclose(cost,cost_new,rtol=tol):
            break
        cost=cost_new
        md=(-2/n)*sum(x*(y-y_predict))
        bd=(-2/n)*sum(y-y_predict)
        m=m-learning_rate*md
        b=b-learning_rate*bd
    return m,b


```

```python
m,b=gradient_descent(np.array(df.iloc[:,1]),np.array(df.iloc[:,2]))
print(f"final m:{m}, b:{b}")
```
